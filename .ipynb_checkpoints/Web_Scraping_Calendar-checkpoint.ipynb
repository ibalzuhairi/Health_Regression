```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

url = "https://hijri-calendar.com/Rabiulakhir/1442"
url = "https://hijri-calendar.com/Jamadilawal/1442"
response = requests.get(url)
soup = BeautifulSoup(response.text, "lxml")
Hijri = []
Melady =[]
cols = ['HijryDate','MeladyDate']

```

```python
for  el in soup.find_all(class_='hijri'):
    Hijri.append(el.text)

skip = 0 
el_list = soup.find_all(class_='meladi')
for el in range(1,len(el_list),2):
    Melady.append(el_list[el].text)
    #print(el_list[el])
    
assert len(Melady) == len(Hijri)
#print(list(zip(Hijri,Melady)))
list(zip(Hijri,Melady))
df = pd.DataFrame(columns=cols,data=list(zip(Hijri,Melady)))
df
```

```python
df["Date"] = pd.to_datetime(df['MeladyDate'])
df["AD_Year"] = df["Date"].dt.year
df["AD_Month"] = df["Date"].dt.month
df["AD_Day"] = df["Date"].dt.day

df[['AH_Year','AH_Month','AH_Day']] = (df['HijryDate'].str.split('/',expand=True)).astype('int')


```
